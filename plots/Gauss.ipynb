```python
from matplotlib import pyplot as plt
import numpy as np
import SimpleITK as sitk
import seaborn as sns
import pandas as pd
%matplotlib inline
```

```python
size = 9
factors = np.zeros((size,size))
factors[size/2,size/2] = 100
#factors[size/2+1,size/2+1] = 100
#factors[size/2-1,size/2-1] = 100
#factors[size/2-1,size/2+1] = 100
#factors[size/2+1,size/2-1] = 100

gaused = sitk.GetArrayFromImage(sitk.DiscreteGaussian(sitk.GetImageFromArray(factors), 1))
gaused_half = sitk.GetArrayFromImage(sitk.DiscreteGaussian(sitk.GetImageFromArray(factors), 0.5))
```

```python
plt.figure(figsize=(5,5))
sns.heatmap(factors.astype(np.uint16), annot=True, cbar=False, fmt="d")
plt.tight_layout()
plt.savefig("before_gauss.png")
```

```python
plt.figure(figsize=(5,5))
sns.heatmap(gaused, annot=True, vmax=100, cbar=False)
plt.tight_layout()
plt.savefig("after_gauss.png")
```

```python
plt.figure(figsize=(5,5))
sns.heatmap(gaused_half, annot=True, vmax=100, cbar=False)
plt.savefig("after_half_gauss.png")
```
